```python
import pandas as pd
import matplotlib.pyplot as plt

import numpy as np
```

```python
import pandas as pd

previous = pd.read_csv("fandango_score_comparison.csv")
after = pd.read_csv("movie_ratings_16_17.csv")


```

```python
fandango_previous = previous[['FILM', 'Fandango_Stars', 'Fandango_Ratingvalue', 'Fandango_votes', 'Fandango_Difference']].copy()
fandango_after = after[['movie', 'year', 'fandango']]
```

```python
#Our goal is to determine whether following Hickey's analysis, the variable Fandango_Stars of fandango_previous has changed
#We are going to do so by comparing the previous and the after samples of data. Previous data - 2015
#After data - 2017
#We can further gather that the 2016-2017 dataset is a sample and only includes films released in 2016-> as well
#as only including films which have a signifcant number of reviews (30)
#
#The sampling technique used is called purposing sampling -- also known as subjective sampling
```

```python
#As our sampling is not completely random and we desire for it to be as representative. We will do as by comparing 
#the 2015 files from the previous dataset to the 2016 films of after dataset
```

```python
fandango_after.sample(10, random_state = 1)
```

```python
sum(fandango_previous['Fandango_votes'] < 30)
```

```python
fandango_previous["Year"] = fandango_previous["FILM"].str[-5:-1] 
```

```python


fandango_previous['Year'].value_counts()

```

```python
fandango_2015 = fandango_previous[fandango_previous["Year"] == "2015"].copy() 
```

```python
fandango_2015
```

```python
fandango_after["year"].value_counts()
```

```python
fandango_2016 = fandango_after[fandango_after["year"] == 2016].copy() 
```

```python
fandingo_2016
```

```python
%matplotlib inline
plt.style.use("fivethirtyeight")

fandango_2016["fandango"].plot.kde(label = "2016", legend=True,figsize=(8,6))
fandango_2015["Fandango_Stars"].plot.kde(label = "2015", legend=True)

plt.title("Kernel Density plot comparing the rating for 2015 and 2016 movies", y = 1.05)

plt.xlabel('Stars')
plt.xlim(0,5) # because ratings start at 0 and end at 5
plt.xticks(np.arange(0,5.1,.5))
plt.show()
```

```python
#As we can see, the 2015 dataset is more skewed to the right than 2016. eg. 6.9% of all rating in 2015 are 5.0 stars 
#whereas, 2016 has 0.05
```

```python
#Additionally, 2015 has a higher density of 4.0
```

```python
print("2015")

fandango_2015["Fandango_Stars"].value_counts(normalize = True).sort_index() * 100
```

```python
print("2016")

fandango_2016["fandango"].value_counts(normalize = True).sort_index() * 100
```

```python
#secondly, 2016's most common rating is 4.0, whereas 2015 is 4.5. The direction is negative from 2015 to 2016
```

```python
mean_2015 = fandango_2015["Fandango_Stars"].mean()
mean_2016 = fandango_2016["fandango"].mean()

median_2015 = fandango_2015["Fandango_Stars"].median()
median_2016 = fandango_2016["fandango"].median()

mode_2015 = fandango_2015["Fandango_Stars"].mode()[0]
mode_2016 = fandango_2016["fandango"].mode()[0]

summary = pd.DataFrame()
summary["2015"] = [mean_2015, median_2015, mode_2015]
summary["2016"] = [mean_2016, median_2016, mode_2016]

summary.index = ['mean', 'median', 'mode']
summary



```

```python
plt.style.use("fivethirtyeight")

summary["2015"].plot.bar(color = "red", align="center", label = "2015", width = 0.25)
summary["2016"].plot.bar(color = "blue", align="edge", label = "2016", width = 0.25, rot = 0, figsize = (8,6))
plt.ylim(0,5.5)
plt.yticks(np.arange(0,5.1,.5))
plt.ylabel("Stars")
plt.title("Comparisson of 2015 vs 2016")

plt.legend(framealpha=0, loc = "upper center")
```

```python
#2016 is slightly lower, with mean (0.2 difference) and mode (0.5). Median is about the same
```

```python
#Conclusion

#Our analysis showed that there is indeed a slight difference between Fandango's ratings for popular movies in 2015 and Fandango's ratings for popular movies in 2016. We also determined that, on average, popular movies released in 2016 were rated lower on Fandango than popular movies released in 2015.

#We cannot be completely sure what caused the change, but the chances are very high that it was caused by Fandango fixing the biased rating system following Hickey's analysis.
```
